```python
df1['DATE'] = pd.to_datetime(df1['DATE'], origin='1899-12-30', unit='D')
df1.head()
```

```python
print("Date range:", df1['DATE'].min(), "to", df1['DATE'].max())
```

```python
import re
df1['PACK_SIZE'] = df1['PROD_NAME'].str.extract(r'(\d+)g', flags=re.IGNORECASE)
df1['PACK_SIZE'] = pd.to_numeric(df1['PACK_SIZE'])
```

```python
df1.head()
```

```python
df1['BRAND'] = df1['PROD_NAME'].str.split().str[0]
```

```python
df1['PROD_DESC'] = df1['PROD_NAME'].str.split(r'\d+[gG]').str[0].str.strip()
```

```python
# merge both csv files into 1 (merge_df)
merged_df = df1.merge(df2, on='LYLTY_CARD_NBR', how='left')
print("\nMerged dataset shape:", merged_df.shape)
print(merged_df.head())
```

```python
merged_df.head()
```

```python
# Total sales by LIFESTAGE and PREMIUM_CUSTOMER
total_sales = merged_df.groupby(
                ['LIFESTAGE', 'PREMIUM_CUSTOMER']
              )['TOT_SALES'].sum().reset_index()
print(total_sales.sort_values('TOT_SALES', ascending=False).head(10))

# Number of unique customers per segment
customer_count = merged_df.groupby(
                    ['LIFESTAGE', 'PREMIUM_CUSTOMER']
                 )['LYLTY_CARD_NBR'].nunique().reset_index()
customer_count.columns = ['LIFESTAGE', 'PREMIUM_CUSTOMER', 'CUSTOMER_COUNT']

#  Merge and calculate average sales per customer
segment_summary = total_sales.merge(
                    customer_count, 
                    on=['LIFESTAGE', 'PREMIUM_CUSTOMER']
                  )
segment_summary['AVG_SALES_PER_CUSTOMER'] = (
    segment_summary['TOT_SALES'] / segment_summary['CUSTOMER_COUNT']
)


print("\nSegment Summary")
print(segment_summary.sort_values('TOT_SALES', ascending=False).to_string())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CHART 1: Total Sales by Segment
plt.figure(figsize=(14, 7))
sns.barplot(
    data=segment_summary.sort_values('TOT_SALES', ascending=False),
    x='LIFESTAGE',
    y='TOT_SALES',
    hue='PREMIUM_CUSTOMER'
)
plt.title('Total Chip Sales by Customer Segment')
plt.xlabel('Lifestage')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart1_total_sales.png')
plt.show()
```

```python
# CHART 2: Average Sales Per Customer
plt.figure(figsize=(14, 7))
sns.barplot(
    data=segment_summary.sort_values('AVG_SALES_PER_CUSTOMER', ascending=False),
    x='LIFESTAGE',
    y='AVG_SALES_PER_CUSTOMER',
    hue='PREMIUM_CUSTOMER'
)
plt.title('Average Chip Spend Per Customer by Segment')
plt.xlabel('Lifestage')
plt.ylabel('Avg Sales Per Customer ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart2_avg_sales.png')
plt.show()
```

```python
# CHART 3: Customer Count by Segment 
plt.figure(figsize=(14, 7))
sns.barplot(
    data=segment_summary.sort_values('CUSTOMER_COUNT', ascending=False),
    x='LIFESTAGE',
    y='CUSTOMER_COUNT',
    hue='PREMIUM_CUSTOMER'
)
plt.title('Number of Customers by Segment')
plt.xlabel('Lifestage')
plt.ylabel('Customer Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart3_customer_count.png')
plt.show()
```

```python
#Average pack size by LIFESTAGE and PREMIUM_CUSTOMER
pack_analysis = merged_df.groupby(
    ['LIFESTAGE', 'PREMIUM_CUSTOMER']
)['PACK_SIZE'].mean().reset_index()
pack_analysis.columns = ['LIFESTAGE', 'PREMIUM_CUSTOMER', 'AVG_PACK_SIZE']
print(pack_analysis.sort_values('AVG_PACK_SIZE', ascending=False).to_string())

# CHART 4: Pack Size Preference
plt.figure(figsize=(14, 7))
sns.barplot(
    data=pack_analysis.sort_values('AVG_PACK_SIZE', ascending=False),
    x='LIFESTAGE',
    y='AVG_PACK_SIZE',
    hue='PREMIUM_CUSTOMER'
)
plt.title('Average Pack Size Preference by Customer Segment')
plt.xlabel('Lifestage')
plt.ylabel('Average Pack Size (g)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart4_pack_size.png')
plt.show()

#Most popular pack sizes overall
print("\nTop 10 Most Popular Pack Sizes:")
print(merged_df['PACK_SIZE'].value_counts().head(10))
```

```python
# SUMMARY INSIGHTS & RECOMMENDATION 
print(" QUANTIUM RETAIL ANALYTICS — CHIP CATEGORY REPORT")

#  KEY METRICS SUMMARY 
print("\n KEY METRICS SUMMARY")
print(f"Total Transactions Analysed : {len(merged_df):,}")
print(f"Total Revenue (Chips)        : ${merged_df['TOT_SALES'].sum():,.2f}")
print(f"Unique Customers             : {merged_df['LYLTY_CARD_NBR'].nunique():,}")
print(f"Date Range                   : {merged_df['DATE'].min().date()} to {merged_df['DATE'].max().date()}")
print(f"Unique Products              : {merged_df['PROD_NAME'].nunique():,}")
print(f"Most Common Pack Size        : {merged_df['PACK_SIZE'].mode()[0]}g")

#TOP 3 SEGMENTS BY TOTAL SALES 
print("\n TOP 3 SEGMENTS BY TOTAL SALES")
top3_sales = segment_summary.sort_values('TOT_SALES', ascending=False).head(3)
for i, row in top3_sales.iterrows():
    print(f"  {row['LIFESTAGE']} — {row['PREMIUM_CUSTOMER']}")
    print(f"    Total Sales      : ${row['TOT_SALES']:,.2f}")
    print(f"    Customer Count   : {int(row['CUSTOMER_COUNT']):,}")
    print(f"    Avg Spend/Customer: ${row['AVG_SALES_PER_CUSTOMER']:.2f}")
    print()

# TOP 3 SEGMENTS BY AVG SPEND PER CUSTOMER 
print("\n TOP 3 SEGMENTS BY AVG SPEND PER CUSTOMER")
top3_avg = segment_summary.sort_values(
               'AVG_SALES_PER_CUSTOMER', ascending=False
           ).head(3)
for i, row in top3_avg.iterrows():
    print(f"  {row['LIFESTAGE']} — {row['PREMIUM_CUSTOMER']}")
    print(f"    Avg Spend/Customer: ${row['AVG_SALES_PER_CUSTOMER']:.2f}")
    print()

# PACK SIZE INSIGHT
print("\n PACK SIZE INSIGHT")
print(f"  Average Pack Size across ALL segments: "
      f"{merged_df['PACK_SIZE'].mean():.1f}g")
print(f"  Min: {merged_df['PACK_SIZE'].min()}g  |  "
      f"Max: {merged_df['PACK_SIZE'].max()}g")
print("  ⚠ Pack size is CONSISTENT across all segments —")
print("    strategy should focus on frequency & basket size,")
print("    NOT pack size variation.")

```
